```python
def calculate_r_2_for_feature(data, feature, regressor = DecisionTreeRegressor()):
    new_data = data.drop(feature, axis=1)

    X_train,X_test,\
    y_train,y_test \
    = train_test_split(new_data,data[feature],test_size=0.25)
    
    if (regressor != DecisionTreeRegressor()):
        scaler = StandardScaler()
        scaler.fit(X_train, y_train)
        X_train = scaler.transform(X_train, y_train)
        X_test = scaler.transform(X_test, y_test)

    regressor.fit(X_train, y_train)

    score = regressor.score(X_test, y_test)
    return score
```

```python
def mean_r2_for_feature(data, feature, method, regressor = DecisionTreeRegressor()):
    scores = []
    for _ in range(100):
        scores.append(method(data, feature, regressor))
        
    scores = np.array(scores)
    return scores.mean()
```
